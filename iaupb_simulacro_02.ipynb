## **Etapa 1**

```python
import pandas as pd

ruta = "/precipitacion_2026enero__Medellin_AeropuertoEnriqueOlayaHerrera.csv"

df = pd.read_csv(ruta, sep=';', decimal =',')

df.head()
```

Verificar que se haya subido el .csv y todos sus datos

```python
df.shape
```

Acá es el **punto 2**: resolver el problema de formatos mixtos de fechas

```python
df["FechaObservacion"] = pd.to_datetime(
    df["FechaObservacion"],
    errors="coerce"
)

df.info()
```

**Punto 3**: Eliminar duplicados exactos (misma fecha/hora y valor)

Primero ejecutare un comando para ver cuantos duplicados tengo

```python
duplicados_exactos = df.duplicated(
    subset=["FechaObservacion", "ValorObservado"]
).sum()

duplicados_exactos
```

Acá eñimino los duplicados exactos, que son 7110. Por lo que usaré el .drop , ya que el drop es parecido al truncate, pero elimina completamente una info especifioca

```python
df = df.drop_duplicates(
    subset=["FechaObservacion", "ValorObservado"]
)
```

Para comprobar que efectivamente el comando anterior funciono, usaré este otro comando y el valor al ejecutar es diferente

```python
df.shape
```

Si funciono ya que los datos iniciales fueron: 24102 - 7110(datos duplicados) = 16992
